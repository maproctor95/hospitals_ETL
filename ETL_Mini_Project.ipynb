```python
# Import dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

### Extract the Hospital.csv Data

```python
# Read the data into a Pandas DataFrame
Hospital_info_df = pd.read_csv('Data/Hospitals_data_world.csv')
Hospital_info_df.head()
```

```python
# Rename the name columns.
Hospital_info_df = Hospital_info_df.rename(columns={'NAME': 'Hospital'})
Hospital_info_df.head()
```

```python
# Get a brief summary 
Hospital_info_df.info()
```

```python
# Get the columns.
Hospital_info_df.columns
```

### Extract the children_hospitals.csv data

```python
# Read the data into a Pandas DataFrame
children_hospital_info_df = pd.read_csv('Data/childrens_hospitals.csv')
children_hospital_info_df.head()
```

```python
# Get a brief summary 
children_hospital_info_df.info()
```

```python
# Get the columns.
children_hospital_info_df.columns
```

```python
children_hospital_info_df[["childrens_hospital"]] = 'TRUE'
```

```python
children_hospital_info_df.head()
```

### Merge Data Frames together

```python
Hospitals_merged_df = Hospital_info_df.merge(children_hospital_info_df, on='Hospital', how='left')

Hospitals_merged_df.tail()
```

```python
# Get the columns.
Hospitals_merged_df.columns
```

```python
#Delete the unwanted columns
Hospitals_merged_df = Hospitals_merged_df.drop(columns={'X', 'Y', 'ID',
       'ZIP4', 
       'COUNTYFIPS', 
       'NAICS_CODE', 'NAICS_DESC', 'SOURCE', 'SOURCE_DAT', 'VAL_METHOD',
       'VAL_DATE', 'ALT_NAME', 'ST_FIPS', 'OWNER',
       'TTL_STAFF', 'DATE_CREAT', 'Beds',
       'Alternative Name', 'Address', 'Address 1', 'Address 2', 'City',
       'State', 'Zip', 'Hospital Website', 'Latitude', 'Longitude',
         'Type', 'Wikipedia Source',
       'Magnet Designation'})
Hospitals_merged_df.head()
```

```python
Hospitals_merged_df.columns
```

```python
# Rename the columns.
Hospitals_merged_df = Hospitals_merged_df.rename(columns={'Hospital': 'hospital_name', 'ADDRESS': 'address', 'CITY': 'city', 'STATE': 'state', 'ZIP': 'zip', 'TYPE': 'type', 'STATUS': 'status',
       'POPULATION': 'population', 'COUNTY': 'county', 'COUNTRY': 'country', 'LATITUDE': 'latitude', 'LONGITUDE': 'longitude', 'WEBSITE': 'website',
       'BEDS': 'beds', 'TRAUMA': 'trauma', 'HELIPAD': 'helipad', 'ADDRESS2': 'address2', 'TELEPHONE':'telephone', 'FID': 'fid', 'STATEID': 'stateid'})
Hospitals_merged_df.head()
```

```python
# Check the datatypes
Hospitals_merged_df.dtypes
```

```python
# Export the DataFrame as a CSV file. 
Hospitals_merged_df.to_csv("Data/Hospitals_merged.csv", index=False)
```
